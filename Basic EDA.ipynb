# EDA (Exploratory Data Analysis):

```python
import pandas as pd
import numpy as np
```

```python
naga = {
    "ID":["101","102","103",np.nan,"105","106","107","108","109","110"],
    "Name": ["MS Dhoni","Virat Kholi","Rohit Sharma","Smriti Mandhana","Harmanpreet Kaur","Mithali Raj","KL Rahul","Ellyse Perry","Lauren bell","Jasprit Bumrah"],
    "Teams": ["CSK","RCB","MI","RCB","MI","GG","DC","RCB","RCB","MI"],
    "Practice_Hours": ["4","8","6","100","6","4","5",np.nan,"2","2"],
    "Gender": ["M","M","M","F","F","F","M","F","F","M"],
    "Playing_Match": ["16","16","15",np.nan,"11","12","10","Fourteen","11","16"],
    "Runs":["350","780","640","450","420","380",np.nan,"Unavailable","98","104"]
}

naga
```

# **CONVERT TO DATAFRAME**

```python
naga = pd.DataFrame(naga)
naga
```

# **CHANGE COLUMN TYPE**

```python
naga.dtypes
```

```python

#  Change the Column type:

naga["ID"] = pd.to_numeric(naga["ID"])
naga["Practice_Hours"] = pd.to_numeric(naga["Practice_Hours"])
naga.dtypes
```

# Replace text with nan:

```python
naga['Playing_Match'] = naga['Playing_Match'].replace('Fourteen', np.nan)
naga["Runs"] = naga["Runs"].replace("Unavailable",np.nan)
```

```python
naga.tail(3)
```

```python
naga['Playing_Match'] = pd.to_numeric(naga['Playing_Match'])
naga['Runs'] = pd.to_numeric(naga['Runs'])

naga.dtypes
```
